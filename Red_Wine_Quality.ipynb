# Feature Selection

Though my target is ***'Qality'*** so drop it from data frame and condier the new dataframe is feature

```python
x = df.drop('quality',axis=1)
y = df['quality']
```

```python
x.shape
```

```python
y.shape
```

**LinearRegression library import**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as TTS
```

```python
LR= LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
LR.fit(x_train,y_train)
```

```python
print(LR.intercept_)
```

```python
prediction_of_system = LR.predict(x_test)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, prediction_of_system))
print('MSE:', metrics.mean_squared_error(y_test, prediction_of_system))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, prediction_of_system)))
```

```python
print('R sq: ',LR.score(x_train,y_train))
```
