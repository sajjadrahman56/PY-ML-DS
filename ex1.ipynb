```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/ex1data1.txt')
```

```python
# here I separate the features and label
#features always 2D where label is 1D ( D = Dimension )

x=df[['populations']]
y=df['profit']
```

```python
plt.scatter(df['populations'],df['profit'])
```

```python
plt.xlabel("populations")
plt.ylabel("profit")
plt.title("populations VS profit")
plt.scatter(x,y)
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x,y,test_size = 0.20,random_state=42)
```

```python
xtest.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg
```

```python
reg.fit(xtrain,ytrain)
```

```python
reg.score(xtest,ytest)
```

```python
x1=df[['populations']]
y1=df['profit']
plt.scatter(x1,y1)
plt.plot(df.populations,reg.predict(df[['populations']]),color="red")
```

```python
reg.predict(xtest)
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
```

```python
y_pred = reg.predict(xtest)
```

```python
m_se = mse(ytest,y_pred)
m_se
```

```python
m_ae=mae(ytest,y_pred)
m_ae
```

```python
rmse = np.sqrt(m_se)
rmse
```
