```python
# Searching Two D array
def search(array,value):
  for i in range(len(array)):
    for j in range(len(array[0])):
      if array[i][j]==value:
        return print(f'value find at index i = [{i}][{j}]')

  return 'no value '

towD = np.array([[1,2,6],[4,50,9],[5,9,8]])
search(towD,50)
```
