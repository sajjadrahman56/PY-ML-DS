```python
# Accessing List

shoppingList = ['milk','mango','butter']
print(shoppingList[2])
```

```python
print('milk' in shoppingList)
```

```python
shoppingList[-1]  # reverse list
```

```python
# Traverse a list
for i in shoppingList:
  print(i)
```

```python
for i in range(len(shoppingList)):
  shoppingList[i] = shoppingList[i]+"+"
  print(shoppingList[i])
```

```python
# update a list

shoppingList[2] = 'sussa' # O(1)
```

```python
for i in shoppingList:
  print(i)
```

```python
# Insert at the Beginning
# Any Given Place
# End of the list
# another list to the list

myList = [3,43,2,3,45]
myList
```

```python
# insert
myList.insert(0,1000)
print(myList)

print('now add in the 4 th pos ')
myList.insert(3,1000)             # O(N)
print(myList)
```

```python
# append - end of the list
myList.append(55)               # O(1)
print(myList)
```

```python
# add another list
newList = [30,40,50,23]
myList.extend(newList)      #O(N)
print(myList)
```

```python
# average finding
total = 0
count = 0

while(True):
  print('Enter the num ')
  inp = input()

  if inp == 'done':break

  val = float(inp)
  total +=val
  count+=1

print(f'average  = {total/count}')
```

```python
# using list to hold the input

total_num = list()
count = 0

while(True):
  print('Enter the num ')
  inp = input()

  if inp == 'done':break

  val = float(inp)
  total_num.append(val)

print(f'average  = {sum(total_num)/len(total_num)}')
```
