## Read write and Show videos from Cameras 

```python
import cv2 as cv
```

```python
cap = cv.VideoCapture(0)
```

```python
while(True):
    ret , frame = cap.read() 
    # read return true if frame available and it store frame variabla
    cv.imshow('frame',frame)
    
    if cv.waitKey(1) & 0xFF  == ord('k'):
        break

cap.release()
cv.destroyAllWindows()
```

### Gray Scale image 

```python
import cv2 as cv
cap = cv.VideoCapture(0)

while(True):
    ret , frame = cap.read() 
    # read return true if frame available and it store frame variabla
    # ret store either true or false 
    
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    cv.imshow('frame',gray)
    
    if cv.waitKey(1) & 0xFF  == ord('k'):
        break

cap.release()
cv.destroyAllWindows()
```

#### Frame width and hight know 

```python
import cv2 as cv
cap = cv.VideoCapture(0)

while(True):
    ret , frame = cap.read() 
    
    print(cap.get(cv.CAP_PROP_FRAME_WIDTH)) # width of the frame 
    print(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
    
    print(cap.get(cv.CAP_PROP_SAR_NUM)) # aspect ratio num/den(num) # official open cv 
    
    gray = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    cv.imshow('frame',gray)
    
    if cv.waitKey(1) & 0xFF  == ord('k'):
        break

cap.release()
cv.destroyAllWindows()
```
