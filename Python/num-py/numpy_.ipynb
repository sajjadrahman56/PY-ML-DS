# Stacking and Spliting of Arrays

```python
arr = np.array([[33,44],[45,55]])
```

```python
arr2 = np.array([(1,2),(9,56)])
```

```python
arr2
```

```python
arr2.shape,type(arr2)
```

```python
v_stack_array = np.vstack((arr,arr2))
```
